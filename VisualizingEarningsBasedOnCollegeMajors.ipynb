```python
recent_grads = recent_grads.dropna()
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Sample_size')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women')
```

```python
recent_grads.plot(x='Major_code', y='Median', kind='scatter', title='Median vs. Major_code')
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r - 1]].plot(kind='hist', rot=30)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r - 3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python
recent_grads.sort_values(by='Total')[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads.sort_values(by='Total')[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

Do students in more popular majors make more money?<br />

```python
recent_grads.sort_values(by='Total')[:10].plot.bar(x='Major', y='Rank', legend=False)
recent_grads.sort_values(by='Total')[-10:].plot.bar(x='Major', y='Rank', legend=False)
```

Do students that majored in subjects that were majority female make more money?<br />

```python
recent_grads.sort_values(by='ShareWomen', ascending=False)[:10].plot.bar(x='ShareWomen', y='Rank', legend=False)
recent_grads.sort_values(by='ShareWomen', ascending=False)[-10:].plot.bar(x='ShareWomen', y='Rank', legend=False)
```

Is there any link between the number of full-time employees and median salary?

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
```
